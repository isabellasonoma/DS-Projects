```python
df = df.reset_index(drop=True)
```

```python
#locations
```

```python
df_locations = df.groupby('Country (HQ)').sum()['Number of Investments']
```

```python
df_locations
```

```python
graph = df_locations.plot.bar(y='Number of Investments', use_index=True)
```

```python
df['Founded Date'] = pd.to_datetime(df.loc[:, 'Founded Date'])
```

```python
df_founded_date = df.set_index('Founded Date')
df_founded_date = df_founded_date.sort_index(ascending=True)
```

```python
df_founded_date.head()
```

```python
graph = df_founded_date.plot.bar(y='Number of Investments', use_index=True,xlabel=None)
```

```python
df['Number of Exits']=df['Number of Exits'].astype(float)
```

```python
graph = df_founded_date.plot.bar(y='Number of Exits', use_index=True,xlabel=None)
```

```python
df['Overall workforce']=df['Overall workforce'].astype(float)
```

```python
df_investor_type = df.groupby('Investor Type').sum()['Number of Investments']
```

```python
graph = df_investor_type.plot.bar(y='Number of Investments', use_index=True)
```

```python
df_investment_stage = df.groupby('Investment Stage').sum()['Number of Investments']
```

```python
graph = df_investment_stage.plot.bar(y='Number of Investments', use_index=True)
```

```python
df_percent = df.groupby('Overall workforce % Female').sum()['Number of Investments']
```

```python
dfpercent['Overall workforce % Female']=dfpercent['Overall workforce % Female'].astype(float)
```

```python
graph = dfpercent.plot.bar(y='Overall workforce % Female', use_index=True)
```

```python
graph = dfpercent.plot.bar(y='Number of Investments', use_index=True)
```
